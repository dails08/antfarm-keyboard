```python
%matplotlib inline
import random
import math
import numpy as np
from pandas import DataFrame, Series
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimage
```

```python
def roll(reds=1, blues=0, blacks=0, verbose=False):
    redOptions = ['m','m','h','h','H','c','c','a']
    blueOptions = ['h','h','c','c','h','h','a','a']
    blackOptions = ['m','h','h','C','h','h','m','C']
    results = []
    
    for i in range(reds):
        results.append(redOptions[random.randint(0, len(redOptions)-1)])
    for i in range(blues):
        results.append(blueOptions[random.randint(0, len(blueOptions)-1)])
    for i in range(blacks):
        results.append(blackOptions[random.randint(0, len(blackOptions)-1)])
    #finalResult = ''.join(sorted(results))
    #if verbose:
    #    print finalResult
    return results

    

    
```

```python
#Reds and Blacks trials
trials = 10000
roll_results = []
outfile = open("rb_dice_results.txt", 'w')
outfile.write("reds,blacks,m,a,h,H,c,C\n")
thisRound = ""
for i in range(5):#reds
    for j in range(5):#blacks
        if i+j==0:
            continue
        for k in range(trials):#trials
            formatted_result=""
            roll_result = roll(i,0,j)
            for result_type in ['m','a','h','H','c','C']:
                formatted_result += ", "+str(len([x for x in roll_result if x == result_type]))
            formatted_result = str(i)+","+str(j)+formatted_result+"\n"
            thisRound += formatted_result
outfile.write(thisRound)
outfile.close()
```

```python
class TIEFighter:
    def __init__(self, toTitle):
        self.kind = "TIE Fighter"
        self.title = toTitle
        self.startingHull = 3
        self.currentHull = 3
        self.acted = False
        
    def squadronAttack(self, swarm=False, howlrunnerBonus=False, verbose=False):
        if(howlrunnerBonus):
            print "Howlrunner! Rolling 4U instead of 3U!"
            initialResult = roll(0,4,0)
        else:
            initialResult = roll(0,3,0)
        if swarm:
            for i in range(len(initialResult)):
                if initialResult[i]=='a' or initialResult[i]=='c':
                    reroll = roll(0,1,0)[0]
                    print "Swarm! Rerolling "+initialResult[i]+" to "+reroll
                    initialResult[i]==reroll
                    break
        return initialResult
                    
    def attackShip(self):
        return roll(0,1,0)
    def defend(self, salvo = ['h'], verbose=False):
        if verbose:
            print "Salvo: " + str(salvo)
        damage = 0
        for char in salvo:
            if char=='h':
                damage += 1
            elif char=='H':
                damage +=2
        self.currentHull -= damage
        if (verbose):
            print self.title + " took "+str(damage)+" damage!"
    def status(self):
        print self.title+"("+self.kind+"):"
        print " Hull: "+str(self.currentHull)+'/'+str(self.startingHull)
        print " Acted: "+str(self.acted)
        
        
class Howlrunner:
    def __init__(self, toTitle = "Howlrunner"):
        self.kind = "TIE Fighter"
        self.title = toTitle
        self.startingHull = 3
        self.currentHull = 3
        self.acted = False
        
        self.canScatter = True
        self.scatterDown = False
        self.canBrace = True
        self.braceDown = False
        
    def clearAccuracies(self):
        self.scatterDown = False
        self.braceDown = False
        
    def resetDefenses(self):
        self.canScatter = True
        self.canBrace = True
        
    def squadronAttack(self, swarm=False, howlrunnerBonus=False, verbose=False):
        initialResult = roll(0,3,0)
        if swarm:
            for i in range(len(initialResult)):
                if initialResult[i]=='a' or initialResult[i]=='c':
                    reroll = roll(0,1,0)[0]
                    print "Swarm! Rerolling "+initialResult[i]+" to "+reroll
                    initialResult[i]==reroll
                    break
        if verbose:
            print "Firing "+str(initialResult)
        return initialResult
                    
    def attackShip(self):
        return roll(0,1,0)
    def defend(self, salvo = ['h'], verbose=False):
        if verbose:
            print "Salvo: " + str(salvo)
        
        damage = 0
        for char in salvo:
            if char=='h':
                damage += 1
            elif char=='H':
                damage +=2
            elif char=='a':
                if self.canBrace and not self.braceDown:
                    print "Howlrunner loses Brace!"
                    self.braceDown = True
                elif self.canScatter and not self.scatterDown:
                    print "Howlrunner loses Scatter!"
                    self.scatterDown = True
        if self.canScatter and not self.braceDown:
            self.canScatter = False
            print "Howlrunner scattered!"
            damage = 0
        elif self.canBrace and not self.braceDown:
            if damage == 1:
                print "Howlrunner won't brace for 1 dmg!"
            elif damage > 0:
                self.canBrace=False
                oldDamage = damage
                damage = int(math.ceil(oldDamage/2.0))
                print "Howlrunner braces and turns "+str(oldDamage)+"dmg into "+ \
                str(damage)+"dmg!"
                
        self.currentHull -= damage
        self.clearAccuracies()

        if (verbose):
            print self.title + " took "+str(damage)+" damage!"
    def status(self):
        print self.title+"("+self.kind+"):"
        print " Hull: "+str(self.currentHull)+'/'+str(self.startingHull)
        print " Acted: "+str(self.acted)
        
        
class XWing:
    def __init__(self, toTitle):
        self.kind = "X-wing"
        self.title = toTitle
        self.startingHull = 5
        self.currentHull = 5
        self.acted = False
        
    def squadronAttack(self, verbose=False):
        initialResult = roll(0,4,0)
        return initialResult
                
    def attackShip(self):
        return roll(0,1,0)
    def defend(self, salvo = ['h'], verbose=False):
        if verbose:
            print "Salvo: " + str(salvo)
        damage = 0
        for char in salvo:
            if char=='h':
                damage += 1
            elif char=='H':
                damage +=2
        self.currentHull-=damage
        if (verbose):
            print self.title + " took "+str(damage)+" damage!"
    def status(self):
        print self.title+"("+self.kind+"):"
        print " Hull: "+str(self.currentHull)+'/'+str(self.startingHull)
        print " Acted: "+str(self.acted)
```
